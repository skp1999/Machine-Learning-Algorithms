## RIDGE REGRESSION REGULARIZATION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Training and Testing Dataset

```python
X_train=pd.read_csv('ML_asgn1-train.csv')
X_test=pd.read_csv('ML_asgn1-test.csv')

x=X_train['Feature']
y=X_train['Label']

a=X_test['Feature']
b=X_test['Label']
```

```python
train_error=[]
test_error=[]
```

```python
m=1000              # Number of training examples 
alpha=0.05          # Learning Rate
iter=2000            # Iteration for gradient descent
```

## Ridge Regression (n=2, lamda=0.25)

```python
## Ridge Regression (n=2, lamda=0.25)

theta0=0.1
theta1=0.1
theta2=0.1

lamda=0.25            # Regularization parameter

j_theta=[]
for i in range(0,iter):
    cost=0
    dtheta0=0
    dtheta1=0
    dtheta2=0
    for j in range(0,m):
        sigma=(theta0+theta1*x[j]+theta2*(x[j]**2)-y[j])
        cost=cost+(sigma**2)
        dtheta0=dtheta0+(sigma)
        dtheta1=dtheta1+(sigma)*x[j]
        dtheta2=dtheta2+(sigma)*(x[j]**2)
        
    cost=cost+lamda*((theta0)**2+(theta1)**2+(theta2)**2)
    cost=cost/(2*m)
    j_theta.append(cost)
    dtheta0=(dtheta0+(lamda*theta0))/m
    dtheta1=(dtheta1+(lamda*theta1))/m
    dtheta2=(dtheta2+(lamda*theta2))/m
    theta0=theta0-(alpha)*dtheta0
    theta1=theta1-(alpha)*dtheta1
    theta2=theta2-(alpha)*dtheta2

## Plot of J vs Iterations

plt.figure()
plt.plot(range(0,iter),np.array(j_theta))
plt.show()

print(theta0, theta1, theta2)

## Plot of Training Data and Model

y_pred_train=[]
for i in range(0,m):
    y_pred_train.append(theta0+theta1*x[i]+theta2*(x[i]**2))

plt.figure()
plt.scatter(x,y)
plt.scatter(x,y_pred_train)

## Train error

sum=0
for i in range(0,200):
    sum=sum+((y[i]-y_pred_train[i])**2)
sum=sum/200

train_error.append(sum)

print("Trained error = ", sum)

## Plot of Test Data and model

y_pred_test=[]
for i in range(0,200):
    y_pred_test.append(theta0+theta1*a[i]+theta2*(a[i]**2))


plt.figure()
plt.scatter(a,b)
plt.scatter(a,y_pred_test)   

## Test error

sum=0
for i in range(0,200):
    sum=sum+((b[i]-y_pred_test[i])**2)
sum=sum/200

test_error.append(sum)

print("Test error = ", sum)
```

```python
## Ridge Regression (n=9, lamda=1.00)

theta0=0.1
theta1=0.1
theta2=0.1
theta3=0.1
theta4=0.1
theta5=0.1
theta6=0.1
theta7=0.1
theta8=0.1
theta9=0.1

lamda=1.00

j_theta=[]
for i in range(0,iter):
    cost=0
    dtheta0=0
    dtheta1=0
    dtheta2=0
    dtheta3=0
    dtheta4=0
    dtheta5=0
    dtheta6=0
    dtheta7=0
    dtheta8=0
    dtheta9=0
    
    for j in range(0,m):
        sigma=(theta0+theta1*x[j]+theta2*(x[j]**2)+theta3*(x[j]**3)+theta4*(x[j]**4)+theta5*(x[j]**5)+theta6*(x[j]**6)+theta7*(x[j]**7)+theta8*(x[j]**8)+theta9*(x[j]**9)-y[j])
        cost=cost+(sigma**2)
        dtheta0=dtheta0+(sigma)
        dtheta1=dtheta1+(sigma)*x[j]
        dtheta2=dtheta2+(sigma)*(x[j]**2)
        dtheta3=dtheta3+(sigma)*(x[j]**3)
        dtheta4=dtheta4+(sigma)*(x[j]**4)
        dtheta5=dtheta5+(sigma)*(x[j]**5)
        dtheta6=dtheta6+(sigma)*(x[j]**6)
        dtheta7=dtheta7+(sigma)*(x[j]**7)
        dtheta8=dtheta8+(sigma)*(x[j]**8)
        dtheta9=dtheta9+(sigma)*(x[j]**9)
        
    cost=cost+lamda*(theta0**2+theta1**2+theta2**2+theta3**2+theta4**2+theta5**2+theta6**2+theta7**2+theta8**2+theta9**2)    
    cost=cost/(2*m)
    j_theta.append(cost)
    dtheta0=(dtheta0+lamda*theta0)/m
    dtheta1=(dtheta1+lamda*theta1)/m
    dtheta2=(dtheta2+lamda*theta2)/m
    dtheta3=(dtheta3+lamda*theta3)/m
    dtheta4=(dtheta4+lamda*theta4)/m
    dtheta5=(dtheta5+lamda*theta5)/m
    dtheta6=(dtheta6+lamda*theta6)/m
    dtheta7=(dtheta7+lamda*theta7)/m
    dtheta8=(dtheta8+lamda*theta8)/m
    dtheta9=(dtheta9+lamda*theta9)/m
    
    theta0=theta0-(alpha)*dtheta0
    theta1=theta1-(alpha)*dtheta1
    theta2=theta2-(alpha)*dtheta2
    theta3=theta3-(alpha)*dtheta3
    theta4=theta4-(alpha)*dtheta4
    theta5=theta5-(alpha)*dtheta5
    theta6=theta6-(alpha)*dtheta6
    theta7=theta7-(alpha)*dtheta7
    theta8=theta8-(alpha)*dtheta8
    theta9=theta9-(alpha)*dtheta9

    
## Plot of J vs Iterations

plt.figure()
plt.plot(range(0,iter),np.array(j_theta))
plt.show()

print(theta0, theta1, theta2, theta3, theta4, theta5, theta6, theta7, theta8, theta9)

## Plot of Training Data and Model

y_pred_train=[]
for i in range(0,m):
    y_pred_train.append(theta0+theta1*x[i]+theta2*(x[i]**2)+theta3*(x[i]**3)+theta4*(x[i]**4)+theta5*(x[i]**5)+theta6*(x[i]**6)+theta7*(x[i]**7)+theta8*(x[i]**8)+theta9*(x[i]**9))

plt.figure()
plt.scatter(x,y)
plt.scatter(x,y_pred_train)

## Train error

sum=0
for i in range(0,200):
    sum=sum+((y[i]-y_pred_train[i])**2)
sum=sum/200

train_error.append(sum)

print("Trained error = ", sum)

## Plot of Test Data and model

y_pred_test=[]
for i in range(0,200):
    y_pred_test.append(theta0+theta1*a[i]+theta2*(a[i]**2)+theta3*(a[i]**3)+theta4*(a[i]**4)+theta5*(a[i]**5)+theta6*(a[i]**6)+theta7*(a[i]**7)+theta8*(a[i]**8)+theta9*(a[i]**9))


plt.figure()
plt.scatter(a,b)
plt.scatter(a,y_pred_test)   

## Test error

sum=0
for i in range(0,200):
    sum=sum+((b[i]-y_pred_test[i])**2)
sum=sum/200

test_error.append(sum)

print("Test error = ", sum)
```
