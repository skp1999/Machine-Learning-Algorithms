## POLYNOMIAL REGRESSION (n=1 to n=9)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the train and test data

```python
X_train=pd.read_csv('ML_asgn1-train.csv')
X_test=pd.read_csv('ML_asgn1-test.csv')
```

```python
x=X_train['Feature']
y=X_train['Label']
plt.scatter(x,y)
plt.show()
```

```python
a=X_test['Feature']
b=X_test['Label']
plt.scatter(a,b)
plt.show()
```

```python
train_error=[]
test_error=[]
```

```python
m=1000              # Number of training examples 
alpha=0.05          # Learning Rate
iter=2000          # Iteration for gradient descent
```

## Degree 1

```python
theta0=0.1
theta1=0.1

j_theta=[]    # Cost function

for i in range(0,iter):
    cost=0
    dtheta0=0
    dtheta1=0
    for j in range(0,m):
        sigma=(theta0+theta1*x[j]-y[j])   
        cost=cost+(sigma)**2                           # Cost function
        dtheta0=dtheta0+(sigma)
        dtheta1=dtheta1+(sigma)*x[j]
    
    cost=cost/(2*m)
    j_theta.append(cost)
    
    theta0=theta0-(alpha)*dtheta0/m                   # Gradient Descent
    theta1=theta1-(alpha)*dtheta1/m
    

## Plot of J vs no. of iterations

plt.figure()
plt.plot(range(0,iter),np.array(j_theta))
plt.show()

print(theta0, theta1)

## Plot of Training Data and Trained Model

y_pred_train=[]
for i in range(0,m):
    y_pred_train.append(theta0+theta1*x[i])

plt.figure()
plt.scatter(x,y)
plt.plot(x,y_pred_train)

## Train error 

sum=0
for i in range(0,200):
    sum=sum+((y[i]-y_pred_train[i])**2)
sum=sum/200

train_error.append(sum)

print("Trained error = ", sum)

## Plot of Test data and Trained Model

y_pred_test=[]
for i in range(0,200):
    y_pred_test.append(theta0+theta1*a[i])


plt.figure()
plt.scatter(a,b)
plt.plot(a,y_pred_test)   

## Test error (Squared error on test data)

sum=0
for i in range(0,200):
    sum=sum+((b[i]-y_pred_test[i])**2)
sum=sum/200

test_error.append(sum)

print("Test error = ", sum)       
```

## Degree 2

```python
theta0=0.1
theta1=0.1
theta2=0.1

j_theta=[]
for i in range(0,iter):
    cost=0
    dtheta0=0
    dtheta1=0
    dtheta2=0
    for j in range(0,m):
        sigma=(theta0+theta1*x[j]+theta2*(x[j]**2)-y[j])
        cost=cost+(sigma**2)
        dtheta0=dtheta0+(sigma)
        dtheta1=dtheta1+(sigma)*x[j]
        dtheta2=dtheta2+(sigma)*(x[j]**2)
        
    
    cost=cost/(2*m)
    j_theta.append(cost)
    
    theta0=theta0-(alpha)*dtheta0/m
    theta1=theta1-(alpha)*dtheta1/m
    theta2=theta2-(alpha)*dtheta2/m

## Plot of J vs Iterations

plt.figure()
plt.plot(range(0,iter),np.array(j_theta))
plt.show()

print(theta0, theta1, theta2)

## Plot of Training Data and Model

y_pred_train=[]
for i in range(0,m):
    y_pred_train.append(theta0+theta1*x[i]+theta2*(x[i]**2))

plt.figure()
plt.scatter(x,y)
plt.scatter(x,y_pred_train)

## Train error

sum=0
for i in range(0,200):
    sum=sum+((y[i]-y_pred_train[i])**2)
sum=sum/200

train_error.append(sum)

print("Trained error = ", sum)

## Plot of Test Data and model

y_pred_test=[]
for i in range(0,200):
    y_pred_test.append(theta0+theta1*a[i]+theta2*(a[i]**2))


plt.figure()
plt.scatter(a,b)
plt.scatter(a,y_pred_test)   

## Test error

sum=0
for i in range(0,200):
    sum=sum+((b[i]-y_pred_test[i])**2)
sum=sum/200

test_error.append(sum)

print("Test error = ", sum)
        
```

## Degree 3
